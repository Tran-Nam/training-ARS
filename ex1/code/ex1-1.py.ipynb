```python
import numpy as np
import matplotlib.pyplot as plt

X = []
Y = []
path = '../ex1data1.txt'
with open(path) as f:
    for line in f:
        line = line.split(',')
        X.append([float(line[0])])
        Y.append([float(line[1])])
X = np.array(X)
Y = np.array(Y)
print(X.shape) #97*1
```

```python
plt.plot(X, Y, 'r+')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
# plt.legend()
plt.show()
```

```python
ones = np.ones((X.shape[0], 1))
X_nor = np.concatenate((ones, X), axis = 1)
print(X_nor.shape)
```

```python
def GD(theta_init, x, y, lr, ite):
    theta = theta_init
    for it in range(ite):
        theta = theta - lr * (x.T.dot(x.dot(theta) - y)) / x.shape[0]
    return theta
        
        
        
```

```python
theta_init = np.zeros((2,1))
ite = 1500
lr = .01
```

```python
w_op = GD(theta_init, X_nor, Y, lr, ite)
print(w_op)
```

```python
plt.plot(X, Y, 'r+', label='Training data')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')


xplot = np.arange(0, 24, .2)
yplot = w_op[0] + xplot * w_op[1]
plt.plot(xplot, yplot, label='Linear Regression')

plt.legend()
plt.axis([4.5,24,-5,25])
plt.show()
```
