```python
import numpy as np 
import re
from nltk.stem.porter import PorterStemmer
```

```python
def processEmail(email_path):
    text = open(email_path).read()
    text = text.lower()
    
    #replace digit by 'number'
    text = re.sub(r'\d+', 'number', text)
    #replace $ by 'dollar'
    text = text.replace('$', 'dollar')
    #replace url by 'httpaddr'
    text = re.sub(r'(http|https)://[^\s]*', 'httpaddr', text)
    #replace spam character
    text = re.sub(r'>|<|<>', '', text)
    #replace email url
    text = re.sub(r'[^\s]+@[^\s]+', 'emailaddr', text)
    
    text = re.sub(r'[^a-zA-Z0-9]', ' ', text)
#     text = re.sub(r'>', '', text)
    
    return text
```

```python
text = processEmail('emailSample1.txt')
text = text.split(' ')
text
```
